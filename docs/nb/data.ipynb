```python
import corner as DFM
# --- plotting ---
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
```

```python
y0, x0 = D.get_subhalos('train', 'mags', snapshot=91)
```

```python
print('%i central galaxies in the training dataset' % (y0.shape[0]/4))
```

```python
ytest, xtest = D.get_subhalos('test', 'mags', snapshot=91)
```

```python
fig = plt.figure(figsize=(6,6))
sub = fig.add_subplot(111)
sub.scatter(y0[:,0], y0[:,1], s=1, label='Training Galaxies')
sub.scatter(ytest[:,0], ytest[:,1], s=3, label='Testing Galaxies')
sub.legend(loc='lower right', handletextpad=0, markerscale=10, fontsize=20)
sub.set_xlabel(r'$\log M_*$', fontsize=25)
sub.set_xlim(9., 12.5)
sub.set_ylabel(r'$\log M_h$', fontsize=25)
sub.set_ylim(9., None)
```
